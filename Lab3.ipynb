```python
image = cv.imread('../images/lr3.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
height, width = image.shape[:2]
center = (width/2, height/2)
 
rotate_matrix = cv.getRotationMatrix2D(center=center, angle=-58, scale=1)
 
image = cv.warpAffine(src=image, M=rotate_matrix, dsize=(width, height))
plt.imshow(image)
```

```python
y = 40
x = 60
h = 250
w = 220

image = image[y:y+h, x:x+w]
plt.imshow(image)
```

```python
image = cv.bitwise_not(image)

plt.imshow(image)
```

```python
gamma_corrected = np.array(255*(image / 255) ** 1.85, dtype = 'uint8')

plt.imshow(gamma_corrected)
```

```python
gaussian = cv.GaussianBlur(image,(5,5),1)

plt.imshow(gaussian)
```

```python
kernel = np.asarray([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])

image_median = cv.medianBlur(gamma_corrected, 3)

filtered_image = cv.filter2D(image_median, -1, kernel)

plt.imshow(filtered_image)
```
